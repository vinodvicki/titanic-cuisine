The function create_Ingredients_df processes a given DataFrame df by transposing it, dropping specific columns, summing the values across rows, filtering out rows with all zero values, sorting the resulting DataFrame by the summed values in descending order, and finally returning this processed DataFrame.

```python
def create_Ingredients_df(df):
    ingredients_df = df.T.drop(['cuisine','Unnamed: 0']).sum(axis=1).to_frame(name='value') #Rename the column after sum operation
    ingredients_df = ingredients_df[(ingredients_df.T !=0 ).any()]
    ingredients_df.sort_values(by ='value', ascending = False, inplace = False)

    return ingredients_df
```

```python
thaia_ingredients_df = create_Ingredients_df(thaia_df)
japanese_ingredients_df = create_Ingredients_df(japanese_df)
indian_ingredients_df = create_Ingredients_df(indian_df)
chinese_ingredients_df = create_Ingredients_df(chinese_df)
korean_ingredients_df = create_Ingredients_df(korean_df)
# creating bar plot
thaia_ingredients_df.head(10).plot(kind='bar', figsize=(10,6))
japanese_ingredients_df.head(10).plot(kind='bar', figsize=(10,6))
indian_ingredients_df.head(10).plot(kind='bar', figsize=(10,6))
chinese_ingredients_df.head(10).plot(kind='bar', figsize=(10,6))
korean_ingredients_df.head(10).plot(kind='bar', figsize=(10,6))
```

```python
feature_df = df2.drop(['cuisine', 'Unnamed: 0','rice','garlic','ginger'], axis= 1)
feature_df.head()
```

```python
# prompt: application of smote

from imblearn.over_sampling import SMOTE

# Assuming 'feature_df' is your feature DataFrame and 'df2['cuisine']' is your target variable
X = feature_df.values
y = df2['cuisine'].values

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to oversample the minority classes
X_resampled, y_resampled = smote.fit_resample(X, y)

# Now you have a balanced dataset in X_resampled and y_resampled
# You can use these for your machine learning models
```
